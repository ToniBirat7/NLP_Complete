````python
from setfit import SetFitModel
````

````python
model = SetFitModel.from_pretrained("BAAI/bge-small-en-v1.5")
````

````python
from datasets import load_dataset

dataset = load_dataset("SetFit/sst2")
dataset
````

````python
# As we will need to have limited data so we'll use sample data
from setfit import sample_dataset

training_data = sample_dataset(dataset['train'], label_column='label', num_samples=10)
training_data
````

````python
# Validation data
val_data = sample_dataset(dataset['validation'], label_column='label', num_samples=5)
val_data
````

````python
test_data = dataset["test"]
test_data
````

````python
# Set the label for the model
model.labels = ["negative", "positive"]
````

````python
# Set the training arguments
from setfit import TrainingArguments

args = TrainingArguments(batch_size=32, num_epochs=10)
````

````python
# Init the trainer
from setfit import Trainer

trainer = Trainer(model=model, args=args, train_dataset=training_data, eval_dataset=test_data )
````

````python
# Train the model
trainer.train()
````

````python
# Metrics
fewshot_metrics = trainer.evaluate()
fewshot_metrics
````

````python
# Save the Model
model.save_pretrained("setfit-bge-small-v1.5-sst2-8-shot")
````
