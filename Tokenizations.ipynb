```python
corpus = """ Hi it's birat. I am 20. I am practicing """
corpus
```

```python
import nltk 
# ensure the tokenizer models are available
nltk.download('punkt', quiet=True)
# some environments expect the 'punkt_tab' tokenizer resource â€” download it as well to avoid LookupError
nltk.download('punkt_tab', quiet=True)

from nltk.tokenize import sent_tokenize, word_tokenize

# split into sentences
token = sent_tokenize(corpus)

# tokenize each sentence into words (result is a list of lists)
token = [word_tokenize(sent) for sent in token]

token
```

```python
# Stemming

words = ['eats', 'eat', 'eating', 'programs', 'programming', 'history']

# PorterStemmer 

from nltk.stem import PorterStemmer

port = PorterStemmer()

for i in words:
  print(port.stem(i))
```

```python
# Lemmatization 
from nltk.stem import WordNetLemmatizer

stm = WordNetLemmatizer()

stm.lemmatize("History", "n")
```
